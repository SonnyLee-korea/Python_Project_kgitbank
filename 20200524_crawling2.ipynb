## requests를 이용한 다이렉트 크롤링
- selenium은 물리 드라이버를 켜서 진행하는 만큼 속도가 느리다는 단점이 있음
- 브라우저 없이 크롤링을 진행할 수 있도록 도와주는 라이브러리requests를 활용하면 
- 훨씬 빠른 속도로 크롤링을 할 수 있음.

```python
# request.get('접근주소')를 이용해 페이지 접근 가능
# 접송가능주소 200, 리다이렉트 301, 접속불능 404 등의 코드로 접속여부 표시
# 소스코드는 requests.get()의 결과를 변수에 저장한 다음
# 저장변수.txt를 통해 얻어올 수 있음.

response = requests.get("https://www.aladin.co.kr/\
shop/common/wbest.aspx?BestType=Bestseller\
&BranchType=1&CID=0&page=1&cnt=1000&SortOrder=1")
text = response.text
soup = BeautifulSoup(text)
```

```python
class_list = []
# 하단에 num1~num40까지 반복해서 들어가도록 .append() 활용
for idx , i in enumerate(np.arange(1,41)):
    if idx == 18:
        class_list.append('num19_line')
    elif idx == 19:
        class_list.append('num20_line')
    else:
        class_list.append('num{}'.format(i))
```

```python
len(soup.find_all('li',class_=['num1','num2','num3']))
```

```python
soup.find_all('li',class_=class_list)[0].find('p','copy').
```

```python
book = soup.find_all('li',class_=class_list)
```

```python
title = []
author = []
price = []
for k in soup.find_all('li',class_=class_list):
    for idx,i in enumerate(k.find_all('p')):
        if idx ==2:
            title.append(i.get_text())
        if idx ==3:
            author.append(i.get_text())
        if idx ==4:
            price.append(i.get_text())
        elif idx>4:
            break
```

```python
import time
from tqdm import tqdm
title = []
author = []
price = []

for idx in tqdm(range(1,251)):
    response = requests.get('http://www.yes24.com/24/category/bestseller?CategoryNumber=001&sumgb=06&fetchSize=40&PageNumber={}'.format(idx))
    time.sleep(1.5)
    text = response.text
    soup = BeautifulSoup(text)
    book = soup.find_all('td','goodsTxtInfo')
    for i in book:
        title.append(i.find('p').find('a').get_text().strip())
        author.append(i.find_all('a')[2].get_text().strip())
        price.append(i.find_all('p')[1].get_text())
```

```python
df = pd.DataFrame({'title':title,
                   'author':author,
                   'price':price})
```

```python
article_list = set()
for idx in range(1,7):
    response = requests.get('https://news.naver.com/main/list.nhn?mode=LS2D&sid2=231&sid1=104&mid=shm&date=20200523&page={}'.format(idx))
    text = response.text
    soup = BeautifulSoup(text)
    dt_list = soup.find_all('dt')
    for item in dt_list:
        article_list.add(item.find('a')['href'])
```
